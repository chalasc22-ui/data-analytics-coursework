```python
import pandas as pd
```

```python
data = pd.read_csv('http://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None)
```

```python
data
```

```python
data.shape
```

```python
data.columns = ['sepal length', 'sepal width', 'petal length', 'petal width', 'class']
```

```python
data.head()
```

```python
data.describe()
```

```python
data.groupby('class').mean()
```

```python
import matplotlib.pyplot as plt

data.boxplot(column='petal length', by='class', figsize=(8, 5))
plt.title('Petal Length by Species')
plt.suptitle('')
plt.ylabel('centimeter')
plt.show()
```

```python
import matplotlib.pyplot as plt

data.boxplot(column='petal length', by='class', figsize=(12, 3))
plt.title('Petal Length by Species')
plt.suptitle('')
plt.ylabel('centimeter')
plt.show()
```

```python
data.select_dtypes(include='number').corr()
```

```python
data.select_dtypes(include='number')
```

```python
data.select_dtypes(exclude='number')
```

```python
import seaborn as sns

plt.figure(figsize=(8, 6))
sns.heatmap(data.select_dtypes(include='number').corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(data=data, x='petal length', y='petal width', hue='class')
plt.title('Petal Length vs Petal Width')
plt.show()
```

```python
sns.pairplot(data, hue='class')
plt.show()
```

```python
data.select_dtypes(include='number').cov()
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(data.select_dtypes(include='number').cov(), annot=True, cmap='coolwarm')
plt.title('Covariance Heatmap')
plt.show()
```
